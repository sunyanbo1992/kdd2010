```python
import tensorflow as tf
import numpy as np
from tensorflow.contrib import rnn
from sklearn.preprocessing import normalize
from sklearn.preprocessing import Normalizer
import csv
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score


from sklearn.metrics import roc_curve, auc

from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report


i = 0
training_set = []
training_label = []

with open('algebra_2005_2006/algebra_2005_2006_train.csv') as csvfile:
    filereader = csv.reader(csvfile, delimiter='\t', quotechar="|")
    
    for row in filereader:
        if i > 0:
            train_data = []
            
            ascii = np.fromstring(row[1], dtype=np.uint8)
            train_data.append(sum(ascii))
                    
            ascii = np.fromstring(row[3] + row[5], dtype=np.uint8)
            train_data.append(sum(ascii))
            
            ascii = np.fromstring(row[-2], dtype=np.uint8)
            train_data.append(sum(ascii))
            
            #train_data = np.asarray(train_data)
            label = np.asarray([0, float(row[13])])
            training_set.append(train_data)
            training_label.append(label)
           
        i += 1
        if i == 9001:
            break
            
testing_set = []
testing_label = []

j = 0
with open('algebra_2005_2006/algebra_2005_2006_master.csv') as csvfile:
    filereader = csv.reader(csvfile, delimiter='\t', quotechar="|")
    
    for row in filereader:
        if j > 0:
            train_data = []
            
            ascii = np.fromstring(row[1], dtype=np.uint8)
            train_data.append(sum(ascii))
                    
            ascii = np.fromstring(row[3] + row[5], dtype=np.uint8)
            train_data.append(sum(ascii))
            
            ascii = np.fromstring(row[-2], dtype=np.uint8)
            train_data.append(sum(ascii))
            
            #train_data = np.asarray(train_data)
            label = np.asarray([0, float(row[13])])
            testing_set.append(train_data)
            testing_label.append(label)
           
        j += 1
        if j == 3001:
            break

training_set = np.asarray(training_set)
testing_set = np.asarray(testing_set)
training_label = np.asarray(training_label)
testing_label = np.asarray(testing_label)

print training_set.shape
print testing_set.shape

transformer = Normalizer().fit(training_set)
training_set = transformer.transform(training_set)
transformer = Normalizer().fit(testing_set)
testing_set = transformer.transform(testing_set)

learning_rate = 0.001
training_steps = 1000
batch_size = 10
display_step = 50

# Network Parameters
num_input = 3
timesteps = 100 # timesteps
num_hidden = 1024 # hidden layer num of features
num_classes = 2

# tf Graph input
X = tf.placeholder("float", [None, timesteps, num_input])
Y = tf.placeholder("float", [None, num_classes])

is_training = tf.placeholder(tf.bool, [])
#seq_len = tf.placeholder(tf.int32, [None])

initializer = tf.contrib.layers.xavier_initializer()
# Define weights
weights = {
    'out': tf.Variable(initializer([num_hidden, num_classes]))
}
biases = {
    'out': tf.Variable(initializer([num_classes]))
}

#inputs = tf.placeholder(tf.float32, [None, None, num_features])

cells = []
for _ in range(2):
    #cell = tf.nn.rnn_cell.GRUCell(num_hidden, activation = 'relu')
    cell = tf.contrib.rnn.LayerNormBasicLSTMCell(num_hidden, activation = tf.nn.relu)
    #cell = BNLSTMCell(num_hidden, training)
    cell = tf.nn.rnn_cell.DropoutWrapper(cell, output_keep_prob=0.8)
    #cell = tf.contrib.rnn.AttentionCellWrapper(cell, attn_length = 60, state_is_tuple=True)
    cells.append(cell)
    
stack = tf.contrib.rnn.MultiRNNCell(cells)
#stack = ExtendedMultiRNNCell(cells, residual_connections = False, residual_combiner="mean")
outputs, _ = tf.nn.dynamic_rnn(stack, X, dtype=tf.float32)

#output_fw, output_bw = outputs

logits = tf.layers.dense(outputs, num_classes)
#logits = tf.matmul(outputs, weights['out']) + biases['out']

#logits = tf.convert_to_tensor(logits,dtype=tf.float32)

#logits = tf.reshape(logits, [-1, num_classes])

prediction = tf.nn.softmax(logits)

# Define loss and optimizer
loss_op = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits, labels=Y))

#loss_op = tf.reduce_mean(tf.square(logits - Y))
#optimizer = tf.train.MomentumOptimizer(0.01, 0.9)
initial_learning_rate = 1e-3
LEARNING_RATE_DECAY_FACTOR = 0.96
global_steps = 1000
DECAY_STEPS = 10000

global_step = tf.Variable(0, trainable=False)
learning_rate = tf.train.exponential_decay(initial_learning_rate,global_step,DECAY_STEPS,
                                           LEARNING_RATE_DECAY_FACTOR,staircase=True)

#optimizer = tf.train.MomentumOptimizer(learning_rate, 0.9).minimize(cost, global_step=global_step)    
    
optimizer = tf.train.RMSPropOptimizer(learning_rate = learning_rate, momentum = 0.9)
#optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)

train_op = optimizer.minimize(loss_op, global_step=global_step)

# Evaluate model (with test logits, for dropout to be disabled)
#correct_pred = tf.equal(tf.argmax(prediction, 1), tf.argmax(Y, 1))

#accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))

# Initialize the variables (i.e. assign their default value)
init = tf.global_variables_initializer()
#init = tf.contrib.layers.xavier_initializer()

# Start training
with tf.Session() as sess:

    sess.run(init)

    for step in range(1, training_steps+1):
        
        for i in xrange(9):
            batch_x = training_set[i*1000:i*1000 + 1000]
            batch_y = training_label[i*1000:i*1000 + 1000]
            batch_x = batch_x.reshape((batch_size, timesteps, num_input))
            
            sess.run(train_op, feed_dict={X: batch_x, Y: batch_y})
            right = sess.run(prediction, feed_dict={X: batch_x, Y: batch_y})

            batch_y = batch_y.reshape(1000, num_classes)
            right = right.reshape(1000, num_classes)
            count = 0
            for j in xrange(len(right)):
                if np.argmax(right[j]) == np.argmax(batch_y[j]):
                    count += 1
            acc = count / 1000.0
            
            test_data = testing_set.reshape(-1, timesteps, num_input)
            
            right_test = sess.run(prediction, feed_dict = {X: test_data})
            right_test = right_test.reshape(3000, num_classes)
            count = 0
            for k in xrange(len(right_test)):
                if np.argmax(right_test[k]) == np.argmax(testing_label[k]):
                    count += 1
            test_acc = count / 3000.0
            
            #acc = sess.run(accuracy, feed_dict={X: batch_x, Y: batch_y})
            if step % display_step == 0 or step == 1:
                loss = sess.run(loss_op, feed_dict={X: batch_x, Y: batch_y})
                print "Step " + str(step) + ", Minibatch Loss= " + "{:.4f}".format(loss) + ", Training Accuracy= " + "{:.3f}".format(acc)
                print "Testing accuracy:", test_acc
                
    print("Optimization Finished!")

    test_data = testing_set.reshape(-1, timesteps, num_input)
    
    print("Testing Accuracy:", sess.run(accuracy, feed_dict={X: test_data, Y: testing_label}))
```
